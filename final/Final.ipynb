```python
import pandas as pd
frame = pd.read_csv("trafficking.csv")
frame = frame.drop(frame.columns[0],1)
frame = frame[frame['citizenship'] == "US"]
frame = frame.reset_index()
frame = frame.drop('index',1)
frame
```

```python
frame["Cases"] = 1
ageFrame = frame.groupby(["majorityStatus","yearOfRegistration"]).sum()
ageFrame['Cases'].plot.bar()
```
